1. Write a "counter" class that can be incremented up to a specified maximum value, will print an error if an attempt is made to increment beyond that value, and allows reseting the counter. 

```python
class Counter:
    def __init__(self, maximum):
        self.maximum = maximum
        self.value = 0
    def increment(self):
        if self.value >= self.maximum:
            print("Error: Maximum reached")
        else:
            self.value += 1
    def reset(self):
        self.value = 0

c = Counter(3)
c.increment()
c.increment()
c.increment()
c.increment()
print(c.value)
c.reset()
print(c.value)
```

```python
import math

class Triangle(Shape):
    def __init__(self, x1,y1,x2,y2,x3,y3):
        self.__points = [(x1,y1),(x2,y2),(x3,y3)]
    def area(self):
        (x1,y1),(x2,y2),(x3,y3)=self.__points
        return abs(x1*(y2-y3)+x2*(y3-y1)+x3*(y1-y2))/2
    def perimeter(self):
        def dist(a,b):
            return math.hypot(a[0]-b[0],a[1]-b[1])
        p=self.__points
        return dist(p[0],p[1])+dist(p[1],p[2])+dist(p[2],p[0])

t = Triangle(0,0,4,0,0,3)
print(t.area(), t.perimeter())
```

7. Add a function to the object classes, including the base, that returns a list of up to 16 pairs of  $x$ and $y$ points on the parameter of the object. 

```python
import math

class Shape:
    def boundary_points(self):
        raise NotImplementedError

class Rectangle(Shape):
    def __init__(self,l,w,x,y):
        self.l=l; self.w=w; self.x=x; self.y=y
    def boundary_points(self):
        pts=[]
        for i in range(4):
            pts.append((self.x,self.y))
        pts=[(self.x,self.y),(self.x+self.l,self.y),(self.x+self.l,self.y+self.w),(self.x,self.y+self.w)]
        return pts

class Circle(Shape):
    def __init__(self,r,x,y):
        self.r=r; self.x=x; self.y=y
    def boundary_points(self):
        pts=[]
        for i in range(16):
            theta=2*math.pi*i/16
            pts.append((self.x+self.r*math.cos(theta),self.y+self.r*math.sin(theta)))
        return pts

print(len(Circle(1,0,0).boundary_points()))
print(Rectangle(2,3,0,0).boundary_points())
```

8. Add a function to the object classes, including the base, that tests if a given set of $x$ and $y$ coordinates are inside of the object. You'll have to think through how to determine if a set of coordinates are inside an object for each object type.

```python
import math

class Rectangle:
    def __init__(self,l,w,x,y):
        self.l=l; self.w=w; self.x=x; self.y=y
    def contains(self,px,py):
        return self.x<=px<=self.x+self.l and self.y<=py<=self.y+self.w

class Circle:
    def __init__(self,r,x,y):
        self.r=r; self.x=x; self.y=y
    def contains(self,px,py):
        return (px-self.x)**2+(py-self.y)**2<=self.r**2

class Triangle:
    def __init__(self,x1,y1,x2,y2,x3,y3):
        self.p=[(x1,y1),(x2,y2),(x3,y3)]
    def contains(self,px,py):
        def area(a,b,c):
            return abs(a[0]*(b[1]-c[1])+b[0]*(c[1]-a[1])+c[0]*(a[1]-b[1]))/2
        A=area(*self.p)
        A1=area((px,py),self.p[1],self.p[2])
        A2=area(self.p[0],(px,py),self.p[2])
        A3=area(self.p[0],self.p[1],(px,py))
        return abs(A-(A1+A2+A3))<1e-6

print(Rectangle(2,3,0,0).contains(1,1))
print(Circle(1,0,0).contains(0.5,0.5))
print(Triangle(0,0,4,0,0,3).contains(1,1))
```

9. Add a function in the base class of the object classes that returns true/false testing that the object overlaps with another object.

```python
class Shape:
    def overlaps(self,other):
        for x,y in self.boundary_points():
            if other.contains(x,y):
                return True
        return False

class Rectangle(Shape):
    def __init__(self,l,w,x,y):
        self.l=l; self.w=w; self.x=x; self.y=y
    def boundary_points(self):
        return [(self.x,self.y),(self.x+self.l,self.y),(self.x+self.l,self.y+self.w),(self.x,self.y+self.w)]
    def contains(self,px,py):
        return self.x<=px<=self.x+self.l and self.y<=py<=self.y+self.w

r1=Rectangle(2,2,0,0)
r2=Rectangle(2,2,1,1)
print(r1.overlaps(r2))
```

10. Copy the `Canvas` class from lecture to in a python file creating a `paint` module. Copy your classes from above into the module and implement paint functions. Implement a `CompoundShape` class. Create a simple drawing demonstrating that all of your classes are working.

```python
class Canvas:
    def __init__(self, width, height):
        self.width = width
        self.height = height
        self.data = [[' '] * width for i in range(height)]

    def set_pixel(self, row, col, char='*'):
        if 0 <= row < self.height and 0 <= col < self.width:
            self.data[row][col] = char

    def get_pixel(self, row, col):
        return self.data[row][col]
    
    def clear_canvas(self):
        self.data = [[' '] * self.width for i in range(self.height)]
    
    def v_line(self, x, y, w, **kargs):
        for i in range(x, x+w):
            self.set_pixel(i, y, **kargs)

    def h_line(self, x, y, h, **kargs):
        for i in range(y, y+h):
            self.set_pixel(x, i, **kargs)
            
    def line(self, x1, y1, x2, y2, **kargs):
        if x1 == x2:
            self.v_line(min(x1,x2), y1, abs(y2-y1)+1, **kargs)
            return
        slope = (y2 - y1) / (x2 - x1)
        for x in range(min(x1,x2), max(x1,x2)+1):
            y = int(y1 + slope * (x - x1))
            self.set_pixel(x, y, **kargs)
            
    def display(self):
        print("\n".join(["".join(row) for row in self.data]))


class Rectangle:
    def __init__(self, x, y, width, height):
        self.x = x
        self.y = y
        self.width = width
        self.height = height

    def draw(self, canvas, char='*'):
        canvas.h_line(self.x, self.y, self.width, char=char)
        canvas.h_line(self.x + self.height - 1, self.y, self.width, char=char)
        canvas.v_line(self.x, self.y, self.height, char=char)
        canvas.v_line(self.x, self.y + self.width - 1, self.height, char=char)


class Line:
    def __init__(self, x1, y1, x2, y2):
        self.x1 = x1
        self.y1 = y1
        self.x2 = x2
        self.y2 = y2

    def draw(self, canvas, char='*'):
        canvas.line(self.x1, self.y1, self.x2, self.y2, char=char)


canvas = Canvas(30, 15)

rect = Rectangle(2, 4, 10, 5)
rect.draw(canvas, char='#')

diag = Line(0, 0, 14, 20)
diag.draw(canvas, char='*')

canvas.display()
```

11. Create a `RasterDrawing` class. Demonstrate that you can create a drawing made of several shapes, paint the drawing, modify the drawing, and paint it again. 

```python
class Rectangle:
    def __init__(self, x, y, width, height, char='*'):
        self.x = x
        self.y = y
        self.width = width
        self.height = height
        self.char = char

    def draw(self, canvas):
        canvas.h_line(self.x, self.y, self.width, char=self.char)
        canvas.h_line(self.x + self.height - 1, self.y, self.width, char=self.char)
        canvas.v_line(self.x, self.y, self.height, char=self.char)
        canvas.v_line(self.x, self.y + self.width - 1, self.height, char=self.char)


class RasterDrawing:
    def __init__(self, width, height):
        self.canvas = Canvas(width, height)
        self.shapes = []

    def add(self, shape):
        self.shapes.append(shape)

    def paint(self):
        self.canvas.clear_canvas()
        for s in self.shapes:
            s.draw(self.canvas)
        self.canvas.display()

    def clear(self):
        self.shapes = []
        self.canvas.clear_canvas()


rd = RasterDrawing(20, 10)
r1 = Rectangle(1, 1, 8, 4)
r2 = Rectangle(3, 10, 6, 3, char='#')

rd.add(r1)
rd.add(r2)
rd.paint()

rd.clear()
rd.paint()
```

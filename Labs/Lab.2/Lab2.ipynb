*Exercise 3:* Modify Exercise 2 so that it takes a matrix in the format from Exercise 1 and draws a tic‑tac‑toe board with `"X"`s and `"O"`s.

````python
# Write your solution here
def draw_tictactoe(board):
    """
    Draws a Tic-Tac-Toe board from a matrix.
    0 = empty, 1 = X, 2 = O
    """
    n = len(board)
    
    def symbol(v):
        return "X" if v == 1 else ("O" if v == 2 else " ")

    horizontal = " ---" * n

    for r in range(n):
        print(horizontal)
        row_str = ""
        for c in range(n):
            row_str += f"| {symbol(board[r][c])} "
        row_str += "|"
        print(row_str)
    print(horizontal)
````

*Exercise 4:* Write a function that takes an `n×n` matrix representing a tic‑tac‑toe game and returns one of the following values:

- `-1` if the game is **incomplete** (still empty spaces and no winner)
- `0` if the game is a **draw**
- `1` if **player 1** (`"X"`) has won
- `2` if **player 2** (`"O"`) has won

Here are some example inputs you can use to test your code:


````python
# Write your solution here
def game_status(board):
    
    n = len(board)

    for r in range(n):
        if board[r][0] != 0 and all(board[r][c] == board[r][0] for c in range(n)):
            return board[r][0]

    for c in range(n):
        if board[0][c] != 0 and all(board[r][c] == board[0][c] for r in range(n)):
            return board[0][c]

    if board[0][0] != 0 and all(board[i][i] == board[0][0] for i in range(n)):
        return board[0][0]

    if board[0][n-1] != 0 and all(board[i][n-1-i] == board[0][n-1] for i in range(n)):
        return board[0][n-1]

    if any(0 in row for row in board):
        return -1

    return 0
````

*Exercise 5:* Write a function that takes a game board, a player number, and `(row, col)` coordinates and places the correct mark (`"X"` or `"O"`) in that location.

Requirements:

- Only allow placing a mark in a previously empty location.
- Return `True` if the move was successful, and `False` otherwise.


````python
# Write your solution here
def place_move(board, player, x, y):
    if x < 0 or y < 0 or x >= len(board) or y >= len(board):
        return False

    if board[x][y] == 0:
        board[x][y] = player
        return True

    return False
````

````python
# Test your solution here
board = [
    [0, 0, 0],
    [0, 0, 0],
    [0, 0, 0]
]

print(place_move(board, 1, 0, 0))  # True
print(place_move(board, 2, 0, 0))  # False (already filled)
print(place_move(board, 2, 1, 1))  # True

for row in board:
    print(row)
````

*Exercise 9:* Use all of the previous exercises to implement a full tic‑tac‑toe game:

- draw the board,
- repeatedly ask two players for a location,
- apply valid moves,
- check the game status until a player wins or the game is a draw.


````python
# Write yourrr solution here
def play_game(board):
    player = 1
 
    while True:
        draw_tictactoe_labeled(board)
        player_turn(board, player)
 
        result = game_status(board)
 
        if result == 1:
            draw_tictactoe_labeled(board)
            print("Player 1 (X) wins!")
            break
        elif result == 2:
            draw_tictactoe_labeled(board)
            print("Player 2 (O) wins!")
            break
        elif result == 0:
            draw_tictactoe_labeled(board)
            print("Game is a draw.")
            break
 
        player = 2 if player == 1 else 1
````

````python
# Test your solution here
def play_tictactoe(n):
    board = [[0]*n for _ in range(n)]
    player = 1

    def draw(board):
        print("   " + "  ".join(str(i+1) for i in range(n)))
        print("  " + " ---"*n)
        for i in range(n):
            row = chr(ord('A')+i) + " "
            for j in range(n):
                if board[i][j] == 1:
                    row += "| X "
                elif board[i][j] == 2:
                    row += "| O "
                else:
                    row += "|   "
            row += "|"
            print(row)
            print("  " + " ---"*n)

    def status(board):
        for i in range(n):
            if board[i][0] != 0 and all(board[i][j] == board[i][0] for j in range(n)):
                return board[i][0]
            if board[0][i] != 0 and all(board[j][i] == board[0][i] for j in range(n)):
                return board[0][i]

        if board[0][0] != 0 and all(board[i][i] == board[0][0] for i in range(n)):
            return board[0][0]
        if board[0][n-1] != 0 and all(board[i][n-1-i] == board[0][n-1] for i in range(n)):
            return board[0][n-1]

        if any(0 in row for row in board):
            return -1
        return 0

    while True:
        draw(board)
        move = input(f"Player {player} move (e.g. A1): ").upper()
        r = ord(move[0]) - ord('A')
        c = int(move[1]) - 1

        if 0 <= r < n and 0 <= c < n and board[r][c] == 0:
            board[r][c] = player
        else:
            print("Invalid move, try again.")
            continue

        result = status(board)
        if result == 1:
            draw(board)
            print("Player 1 (X) wins!")
            break
        elif result == 2:
            draw(board)
            print("Player 2 (O) wins!")
            break
        elif result == 0:
            draw(board)
            print("Game is a draw.")
            break

        player = 2 if player == 1 else 1

play_tictactoe(5)
````

*Exercise 11:* Develop a version of the game where one player is the computer.

Note: you do **not** need an extensive search for the best move. For example, you can have the computer:
- block obvious losses
- otherwise try to create a winning row/column/diagonal


````python
# Write your solution here
def play_vs_computer(n):
    board = [[0]*n for _ in range(n)]

    def draw():
        print("   " + "  ".join(str(i+1) for i in range(n)))
        print("  " + " ---"*n)
        for i in range(n):
            row = chr(ord('A')+i) + " "
            for j in range(n):
                row += "| X " if board[i][j]==1 else "| O " if board[i][j]==2 else "|   "
            print(row+"|")
            print("  " + " ---"*n)

    def status():
        for i in range(n):
            if board[i][0]!=0 and all(board[i][j]==board[i][0] for j in range(n)):
                return board[i][0]
            if board[0][i]!=0 and all(board[j][i]==board[0][i] for j in range(n)):
                return board[0][i]
        if board[0][0]!=0 and all(board[i][i]==board[0][0] for i in range(n)):
            return board[0][0]
        if board[0][n-1]!=0 and all(board[i][n-1-i]==board[0][n-1] for i in range(n)):
            return board[0][n-1]
        if any(0 in row for row in board):
            return -1
        return 0

    def computer_move():
        for player in [2, 1]:
            for i in range(n):
                for j in range(n):
                    if board[i][j] == 0:
                        board[i][j] = player
                        if status() == player:
                            board[i][j] = 2
                            return
                        board[i][j] = 0
        for i in range(n):
            for j in range(n):
                if board[i][j] == 0:
                    board[i][j] = 2
                    return

    while True:
        draw()
        move = input("Your move (e.g. A1): ").upper()
        r = ord(move[0]) - ord('A')
        c = int(move[1]) - 1

        if 0 <= r < n and 0 <= c < n and board[r][c] == 0:
            board[r][c] = 1
        else:
            print("Invalid move")
            continue

        if status() == 1:
            draw(); print("You win!")
            break

        computer_move()

        if status() == 2:
            draw(); print("Computer wins!")
            break
        if status() == 0:
            draw(); print("Draw!")
            break
````

*Exercise 12:* Develop a version of the game where one player is the computer. This time, write a computer player using exhaustive search with a max depth parameter, similar to lecture.

````python
# Write your solution here
def play_vs_computer_minimax(n, max_depth=4):
    board = [[0]*n for _ in range(n)]

    def draw():
        print("  " + " ---"*n)
        for i in range(n):
            row = ""
            for j in range(n):
                row += "| X " if board[i][j]==1 else "| O " if board[i][j]==2 else "|   "
            print(row + "|")
            print("  " + " ---"*n)

    def status():
        for i in range(n):
            if board[i][0]!=0 and all(board[i][j]==board[i][0] for j in range(n)):
                return board[i][0]
            if board[0][i]!=0 and all(board[j][i]==board[0][i] for j in range(n)):
                return board[0][i]
        if board[0][0]!=0 and all(board[i][i]==board[0][0] for i in range(n)):
            return board[0][0]
        if board[0][n-1]!=0 and all(board[i][n-1-i]==board[0][n-1] for i in range(n)):
            return board[0][n-1]
        if any(0 in r for r in board):
            return -1
        return 0

    def minimax(depth, maximizing):
        s = status()
        if s != -1 or depth == 0:
            return 1 if s==2 else -1 if s==1 else 0

        best = -999 if maximizing else 999
        for i in range(n):
            for j in range(n):
                if board[i][j]==0:
                    board[i][j] = 2 if maximizing else 1
                    val = minimax(depth-1, not maximizing)
                    board[i][j] = 0
                    best = max(best,val) if maximizing else min(best,val)
        return best

    def computer_move():
        best, move = -999, None
        for i in range(n):
            for j in range(n):
                if board[i][j]==0:
                    board[i][j]=2
                    val = minimax(max_depth-1, False)
                    board[i][j]=0
                    if val > best:
                        best, move = val, (i,j)
        board[move[0]][move[1]] = 2

    while True:
        draw()
        move = input("Your move (row col): ").split()
        r,c = int(move[0]), int(move[1])
        if board[r][c]!=0: continue
        board[r][c]=1
        if status()==1: draw(); print("You win"); break
        computer_move()
        if status()==2: draw(); print("Computer wins"); break
        if status()==0: draw(); print("Draw"); break
````

*Exercise 13:* Make the 2 computer players play each-other for 10 games on a 3x3, then 4x4, then 5x5 grid. Set the max depth so that the games only take seconds. Measure the "smarter" player's win rate for each grid.

````python
# Write your solution here
import random

def simulate_games(n, games=10, max_depth=3):
    smart_wins = 0

    def empty_board():
        return [[0]*n for _ in range(n)]

    def status(b):
        for i in range(n):
            if b[i][0]!=0 and all(b[i][j]==b[i][0] for j in range(n)):
                return b[i][0]
            if b[0][i]!=0 and all(b[j][i]==b[0][i] for j in range(n)):
                return b[0][i]
        if b[0][0]!=0 and all(b[i][i]==b[0][0] for i in range(n)):
            return b[0][0]
        if b[0][n-1]!=0 and all(b[i][n-1-i]==b[0][n-1] for i in range(n)):
            return b[0][n-1]
        if any(0 in r for r in b):
            return -1
        return 0

    def minimax(b, depth, maxp):
        s = status(b)
        if s != -1 or depth == 0:
            return 1 if s==2 else -1 if s==1 else 0
        best = -999 if maxp else 999
        for i in range(n):
            for j in range(n):
                if b[i][j]==0:
                    b[i][j] = 2 if maxp else 1
                    v = minimax(b, depth-1, not maxp)
                    b[i][j] = 0
                    best = max(best,v) if maxp else min(best,v)
        return best

    def smart_move(b):
        best, move = -999, None
        for i in range(n):
            for j in range(n):
                if b[i][j]==0:
                    b[i][j]=2
                    v = minimax(b, max_depth-1, False)
                    b[i][j]=0
                    if v > best:
                        best, move = v, (i,j)
        b[move[0]][move[1]] = 2

    def random_move(b, player):
        moves = [(i,j) for i in range(n) for j in range(n) if b[i][j]==0]
        i,j = random.choice(moves)
        b[i][j] = player

    for _ in range(games):
        b = empty_board()
        while True:
            random_move(b,1)
            if status(b)==1 or status(b)==0: break
            smart_move(b)
            if status(b)==2:
                smart_wins += 1
                break
            if status(b)==0: break

    return smart_wins / games
````

````python
# Test your solution here
for size in [3, 4, 5]:
    rate = simulate_games(size, games=10, max_depth=3)
    print(f"{size}x{size} board: Smart player win rate = {rate}")
````
